```python
import os
import cv2
import pytesseract
from PIL import Image

# Set Tesseract path (Windows only)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Path to dataset folder
dataset_path =r"D:\Sem_4\Project Based Learning - 1 (PBL1)\archive\SROIE2019\test\img"

# Iterate through all images in the dataset folder
for image_file in os.listdir(dataset_path):
    if image_file.endswith(('.png', '.jpg', '.jpeg')):  # Ensure valid image format
        image_path = os.path.join(dataset_path, image_file)
        
        # Load the image
        img = cv2.imread(image_path)

        # Convert image to grayscale (improves OCR accuracy)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Apply thresholding
        _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # OCR text extraction
        extracted_text = pytesseract.image_to_string(thresh)

        # Print or save results
        print(f"Extracted Text from {image_file}:")
        print(extracted_text)
        print("-" * 50)
```

```python
# Path to save the OCR results
output_file_path = r"D:\Sem_4\Project Based Learning - 1 (PBL1)\archive\SROIE2019\test\ocr_results.txt"

# Open the output file in write mode
with open(output_file_path, 'w') as output_file:
    # Iterate through all images in the dataset folder
    for image_file in os.listdir(dataset_path):
        if image_file.endswith(('.png', '.jpg', '.jpeg')):  # Ensure valid image format
            image_path = os.path.join(dataset_path, image_file)
            
            # Load the image
            img = cv2.imread(image_path)

            # Convert image to grayscale (improves OCR accuracy)
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

            # Apply thresholding
            _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

            # OCR text extraction
            extracted_text = pytesseract.image_to_string(thresh)

            # Write results to the file
            output_file.write(f"Extracted Text from {image_file}:\n")
            output_file.write(extracted_text)
            output_file.write("\n" + "-" * 50 + "\n")

print("OCR results saved to", output_file_path)
```
